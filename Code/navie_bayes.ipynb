```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
```

```python
data_frame = pd.read_csv('train.csv')
data_frame.info()
```

```python
X = data_frame.iloc[:,:-1].values
y = data_frame.iloc[:,-1].values
print(X)
print(y)
```

```python
convert = preprocessing.MinMaxScaler()
X = convert.fit_transform(X)
print(X)
```

```python
np.random.seed(42)
xtrain, xtest, ytrain, ytest = train_test_split(X,y,test_size = 0.2)
```

```python
print('Dữ liệu huấn luyện')
print(xtrain)
print()
print(ytrain)
```

```python
print('Dữ liệu thử nghiệm ban đầu')
print(xtest)
print(ytest)
```

```python
model = GaussianNB()
model.fit(xtrain,ytrain)
ypred = model.predict(xtest)
print(ypred)
```

```python
cm = confusion_matrix(ytest,ypred)
print("Confushion Matrix")
print(cm)
print(f'\nTrue Positives (TP):   {cm[0,0]}')
print(f'False Positives (FP):  {cm[0,1]}')
print(f'False Negatives (FN):  {cm[1,0]}')
print(f'True Negatives (TN):   {cm[1,1]}')
```

```python
print(accuracy_score(ytest,ypred))
```
